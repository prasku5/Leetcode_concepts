```python
# You are given an integer array height of length n. There are n vertical lines drawn such that the two endpoints 
# of the ith line are (i, 0) and (i, height[i]).

# Find two lines that together with the x-axis form a container, such that the container contains the most water.

# Return the maximum amount of water a container can store.

# Notice that you may not slant the container.
```

```python
# Input: height = [1,8,6,2,5,4,8,3,7]
# Output: 49
# Explanation: The above vertical lines are represented by array [1,8,6,2,5,4,8,3,7]. In this case, the max area of 
# water (blue section) the container can contain is 49.
# Example 2:

# Input: height = [1,1]
# Output: 1
```

```python
# Approach

# Simply, we calculate max area of rectangle then return it. Formula is

# width * height
# Let's think about width and height with this example.

# Input: height = [1,8,6,2,5,4,8,3,7]
# Each number is height, so we can easily get height. But how about width?

# width is also simple. It's just distance between two heights. That's why it's good idea to have two pointers left
#  and right.

#  0,1,2,3,4,5,6,7,8 (= index)
# [1,8,6,2,5,4,8,3,7]
#  L               R

# Left pointer starts from index 0
# Right pointer starts from the last index
# Let's see how it works!

# width = right - left = 8 - 0 = 8
# For height, we take smaller height between left and right, because if we calculate based on the taller height, the
#  water would overflow from the container.

# height = min(height[left], height[right]) = min(1, 7) = 1
# Current max area should be

# 8 * 1 = 8
# After that, we want to move one of the pointers. How can we judge it? It's simple. We want to keep taller height 
# between left and right because there is a possibility that we will get max area with the taller height.

# In this case,

# left vs right = 1 vs 7
# So we move the left pointer to next.

#  0,1,2,3,4,5,6,7,8 (= index)
# [1,8,6,2,5,4,8,3,7]
#    L             R

# max_area = 8
# We will repeat the same process until we break left < right condition. I'll speed up.

#  0,1,2,3,4,5,6,7,8 (= index)
# [1,8,6,2,5,4,8,3,7]
#    L             R

# max_area = 8
# current_area = 7 * 7 = 49
# max_area = max(8, 49) = 49
# Move R to next
#  0,1,2,3,4,5,6,7,8 (= index)
# [1,8,6,2,5,4,8,3,7]
#    L           R

# max_area = 49
# current_area = 6 * 3 = 18
# max_area = max(49, 18) = 49
# Move R to next
#  0,1,2,3,4,5,6,7,8 (= index)
# [1,8,6,2,5,4,8,3,7]
#    L         R

# max_area = 49
# current_area = 5 * 8 = 40
# max_area = max(49, 40) = 49
# Move L to next 
# In the above case, we can also move R instead of L because L and Rare the same(= 8).

#  0,1,2,3,4,5,6,7,8 (= index)
# [1,8,6,2,5,4,8,3,7]
#      L       R

# max_area = 49
# current_area = 4 * 6 = 14
# max_area = max(49, 24) = 49
# Move L to next 
#  0,1,2,3,4,5,6,7,8 (= index)
# [1,8,6,2,5,4,8,3,7]
#        L     R

# max_area = 49
# current_area = 3 * 2 = 6
# max_area = max(49, 6) = 49
# Move L to next 
#  0,1,2,3,4,5,6,7,8 (= index)
# [1,8,6,2,5,4,8,3,7]
#          L   R

# max_area = 49
# current_area = 2 * 5 = 10
# max_area = max(49, 10) = 49
# Move L to next 
#  0,1,2,3,4,5,6,7,8 (= index)
# [1,8,6,2,5,4,8,3,7]
#            L R

# max_area = 49
# current_area = 1 * 4 = 4
# max_area = max(49, 4) = 49
# Move L to next 
# Now L and R are the same index. We shop iteration.

#  0,1,2,3,4,5,6,7,8 (= index)
# [1,8,6,2,5,4,8,3,7]
#              L
#              R 
# return 49
```

```python
from typing import List

# we will use two pointers to solve this problem

# Time complexity is O(n) and space complexity is O(1)

# The intuition behind this approach is that the area formed between the lines will always be
# limited by the height of the shorter line.

# At the beginning we consider the area constituting the exterior most lines. Now, to maximize the area, we need to
# consider the area between the lines of larger lengths. If we try to move the pointer at the longer line inwards, we
# won't gain any increase in area, since it is limited by the shorter line. But moving the shorter line's pointer could
# turn out to be beneficial, as per the same argument, despite the reduction in the width. This is done since a relatively

# case1: if the height of the left pointer is less than the height of the right pointer we will move the left pointer
# case2: if the height of the left pointer is greater than the height of the right pointer we will move the right pointer
# case3: if the height of the left pointer is equal to the height of the right pointer we will move the left pointer
# case 3 is because we can move either the left pointer or the right pointer it doesn't matter

# Reason why we do while left < right is because we want to make sure that the left pointer is less than the right pointer
# what is the left most pointer is max at the beginning then we will not be able to calculate the area this is because
# the area is limited by the height of the shorter line

# ---------------
#               |
# |             |(max height at the end)
# ---------------
# L             R

# ---------------
#               |
# |             |(max height at the end) ---> Move the left pointer sinc ethe height of the left pointer is less than the right pointer
# ---------------
# L ->          R

# ---------------
#   |           |
# | |           |(max height at the end)
# ---------------
#   L           R


#If it didnt get into the loop, it means the left pointer is greater than the right pointer and the area is limited by the
# height of the shorter line

class Solution:
    def maxArea(self, height: List[int]) -> int:
        max_area = 0 # we will store the max area here
        left = 0 # we are initializing the left pointer to 0
        right = len(height) - 1 # we are initializing the right pointer to the last index of the height array
        while left < right: # As long as the left pointer is less than the right pointer we will continue the loop
            # we will calculate the area of the container and store the max area
            # we will move the pointer that has the smaller height
            # if the height of the left pointer is less than the height of the right pointer we will move the left pointer
            max_area = max(max_area, min(height[left], height[right]) * (right - left))
            if height[left] < height[right]:
                left += 1
            else:
                right -= 1
        return max_area
        
```
